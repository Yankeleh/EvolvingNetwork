Test 1: Make sure the module loads

```julia
push!(LOAD_PATH, dirname(dirname(@__FILE__)) * "\\src")

using EvolvingNetwork
using Revise
using Test

# Test 1: Module loads
@test EvolvingNetwork !== nothing
```

```julia
# Test 2: Network creation
network = EvolvingNetwork.Network(50)
@test network.net_size == 50
@test size(network.positions) == (50, 3)
```

```julia
# Test 3: Distance computation
pos1 = [0.0, 0.0, 20.0]
pos2 = [0.0, 1.0, 20.0]
dist = EvolvingNetwork.distance(pos1, pos2)

#1 degre longitude at equator (seal level) ~111.32 km
@test 110 < dist < 112
```

```julia
# Test 4: Adjacency construction
using LinearAlgebra
using SparseArrays

network = EvolvingNetwork.Network(100)
EvolvingNetwork.build_adjacency!(network)
@test issymmetric(network.adjacency)
@test nnz(network.adjacency) > 0
```

```julia
# Test 5 - End-to-end simulation
network = EvolvingNetwork.Network(300)
EvolvingNetwork.simulate!(network,1.,30,0.036)  # 1 hour, 30-min timesteps

@test length(network.spectral_history.timestamps) == 2
@test length(network.spectral_history.num_components) == 2
@test all(network.spectral_history.num_components .>= 1)
@test all(0 .<= network.spectral_history.cheeger_lower_bound)
```
